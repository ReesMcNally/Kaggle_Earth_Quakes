```python
import numpy as np 
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline

from os import listdir

from math import sqrt #test
```

```python
proj_dir="C:\\Users\Rees\\Documents\\Python Scripts\\Kaggle\\Earth Quake Guessing\\all (1)\\"
train = pd.read_csv(proj_dir+"train.csv", nrows=400000000)
train.head(2)
```

```python
eventsc=np.append([0],np.argwhere(np.diff(train.time_to_failure.values)>0))
eventsc
```

```python
# This shows the first XX data points, way far zoomed out
fig, ax = plt.subplots(2,1,figsize=(10,9))
ax[0].plot(train.index.values[0:eventsc[3]], train.acoustic_data.values[0:eventsc[3]], c="Red")
ax[0].set_xlabel("Index")
ax[0].set_ylabel("Signal")
ax[0].set_title("Full Signal")


ax[1].plot(train.index.values[0:eventsc[3]], train.time_to_failure.values[0:eventsc[3]])
ax[1].set_xlabel("Index")
ax[1].set_ylabel("time to failure difference")
ax[1].set_title("T2F");
```

```python
# This shows data localized around event 1
trial=1
gap=1E7
cmin=int(eventsc[trial]-gap/2)
cmax=int(eventsc[trial]+gap/2)

fig, ax = plt.subplots(2,1,figsize=(10,9))
ax[0].plot(train.index.values[cmin:cmax], train.acoustic_data.values[cmin:cmax], c="Red")
ax[0].set_xlabel("Index")
ax[0].set_ylabel("Signal")
ax[0].set_title("Failure Zoom")

ax[1].plot(train.index.values[cmin:cmax], train.time_to_failure.values[cmin:cmax])
ax[1].set_xlabel("Index")
ax[1].set_ylabel("time to failure difference")
ax[1].set_title("T2F");
```
